```python
# cve_mun_ls, year_col_ls, quarter_col_ls = [], [], []
cve_mun_hash = {}

for index, row in bosch_rollout.iterrows():
  year = row["year"]
  quarter = row["quarter"]
  cve_mun = row["cvemun"]
  fam_enrolled = row["fam"] #change indv
  if cve_mun not in cve_mun_hash:
    if fam_enrolled > 9:
        cve_mun_hash[cve_mun] = [year, quarter]
        
```

```python
def output_cve_mun(estado, municipio):
  estado = str(int(estado))
  municipio = str(int(municipio))

  if len(municipio) < 2:
    municipio = "00" + municipio
  if len(municipio) < 3:
    municipio = "0" + municipio
  
  return int(estado + municipio)

# stripped_folio = []

#cve_mun = []
household02_tocvemun_dict = {}

#TODO: fix this since only takes into account 02? rethink
for index, row in mflx_mun_02.iterrows():
  if not pd.isna(row['folio']):
    trimmed_folio = int(str(int(row['folio']))[:-3])
    get_cve_mun = output_cve_mun(row['ent'], row['mpio'])
    household02_tocvemun_dict[trimmed_folio] = get_cve_mun
```

```python
mflx_conpor_05['trimmed_folio'] = mflx_conpor_05['folio'].apply(lambda x: int(str(x)[:-3]) if not pd.isna(x) else np.nan)

folio_to_time_interview = {}
for index, row in mflx_conpor_05.iterrows():
  folio = row["trimmed_folio"]
  year = row["anio"]
  month = row["mes"]
  if folio not in folio_to_time_interview:
    folio_to_time_interview[folio] = [year, month]

interview_year_ls = []
interview_month_ls = []
for index, row in households_data_df.iterrows():
  folio = row["trimmed_folio"]
  if folio in folio_to_time_interview:
    year = folio_to_time_interview[folio][0]
    month = folio_to_time_interview[folio][1]
    interview_year_ls.append(year)
    interview_month_ls.append(month)
  else:
    interview_year_ls.append(np.NaN)
    interview_month_ls.append(np.NaN)

households_data_df["interview_yr"] = interview_year_ls
households_data_df["interview_month"] = interview_month_ls
households_data_df.dropna(subset=['interview_yr'], inplace=True)

households_data_df
```
