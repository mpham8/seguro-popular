```python
import pandas as pd
import numpy as np
```

```python
mflx_control_02 = pd.read_stata("hh02dta_all/hh02dta_bc/c_ls.dta")
mflx_control_02
```

```python
mflx_control_02['trimmed_folio'] = mflx_control_02['folio'].apply(lambda x: int(str(int(x))[:-3]) if not pd.isna(x) else np.nan)
mflx_control_02
```

individual stats

```python
#ls14 - education
#secondary 4-10
#high school 6-10
#college 9-10

#ls13_2 - income

#ls04 = sex

#has age - ls02_1 - 1 = age, 2 = month, 3 = days
#age num - ls02_2


#TO DO
#secondary beyond education
#high school beyond education
#college beyond education

def assign_secondary(value):
    if 4 <= value <= 10:
        return 1
    elif 1 <= value <= 3:
        return 0
    else:
        return np.nan
    
def assign_highschool(value):
    if 6 <= value <= 10:
        return 1
    elif 1 <= value <= 5:
        return 0
    else:
        return np.nan
    
def assign_college(value):
    if 9 <= value <= 10:
        return 1
    elif 1 <= value <= 8:
        return 0
    else:
        return np.nan

mflx_control_02['secondary'] = mflx_control_02['ls14'].apply(assign_secondary)
mflx_control_02['highschool'] = mflx_control_02['ls14'].apply(assign_highschool)
mflx_control_02['college'] = mflx_control_02['ls14'].apply(assign_college)


#age

age_yrs_ls = []
for index, row in mflx_control_02.iterrows():
    age_measurement = row["ls02_1"]
    age_units = row["ls02_2"]

    if np.isreal(age_units):
        if age_measurement == 1:
            age_yrs_ls.append(age_units)
        if age_measurement == 2:
            age_yrs_ls.append(1)
        if age_measurement == 3:
            age_yrs_ls.append(1)
        if age_measurement ==8:
            age_yrs_ls.append(np.NaN)
    else:
        age_yrs_ls.append(np.NaN)

mflx_control_02["age_yrs"] = age_yrs_ls
        

#city size


mflx_control_02
```

household stats

```python
#ls05_1 - 1 = household head

#household income - ls13_2
#female head (binary) - ls04
#amount of children
#education of head (head hs, head sec, head college)
#household size

households = mflx_control_02['trimmed_folio'].unique()
household_income_ls = []
household_head_female_ls = []
household_head_secondary_ls = []
household_head_highschool_ls = []
household_head_college_ls = []
household_children_ls = []
household_size_ls = []

# households = [5173]
for household in households:
  household_rows = mflx_control_02[mflx_control_02['trimmed_folio'] == household]
  household_size = len(household_rows)
  household_size_ls.append(household_size)

  household_children_rows = household_rows[household_rows['age_yrs'] < 18]
  household_children = len(household_children_rows)
  household_children_ls.append(household_children)

  household_income = household_rows['ls13_2'].sum()
  household_income_ls.append(household_income)
  # print(household)
  household_head = household_rows[household_rows['ls05_1'] == 1]
  if len(household_head) > 0:
    if household_head.iloc[0]["ls04"] == 3:
      household_head_female_ls.append(1)
    else:
      household_head_female_ls.append(0)

    if household_head.iloc[0]["secondary"] == 1:
      household_head_secondary_ls.append(1)
    elif household_head.iloc[0]["secondary"] == 0:
      household_head_secondary_ls.append(0)
    else:
      household_head_secondary_ls.append(np.NaN)
    
    if household_head.iloc[0]["highschool"] == 1:
      household_head_highschool_ls.append(1)
    elif household_head.iloc[0]["highschool"] == 0:
      household_head_highschool_ls.append(0)
    else:
      household_head_highschool_ls.append(np.NaN)

    if household_head.iloc[0]["college"] == 1:
      household_head_college_ls.append(1)
    elif household_head.iloc[0]["college"] == 0:
      household_head_college_ls.append(0)
    else:
      household_head_college_ls.append(np.NaN)
    
  else:
      household_head_female_ls.append(np.NaN)
      household_head_secondary_ls.append(np.NaN)
      household_head_highschool_ls.append(np.NaN)
      household_head_college_ls.append(np.NaN)



# households
# household_income_ls = []
# household_head_female_ls = []
# household_head_secondary_ls = []
# household_head_highschool_ls = []
# household_head_college_ls = []
# household_children_ls = []
# household_size_ls = []
```
