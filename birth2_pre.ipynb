```python
import pandas as pd
import numpy as np
```

```python
mflx_kid_02 = pd.read_stata("hh02dta_all/hh02dta_b4/iv_ac.dta")
mflx_kid_05 = pd.read_stata("hh05dta_all/hh05dta_b4/iv_ac.dta")
mflx_kid_09 = pd.read_stata("hh09dta_all/hh09dta_b4/iv_ac.dta")
mflx_kid_09
```

```python
mflx_labor_02 = pd.read_stata("mflx_labor_02_edited.dta")
mflx_labor_05 = pd.read_stata("mflx_labor_05_edited.dta")
mflx_labor_09 = pd.read_stata("mflx_labor_09_edited.dta")
mflx_labor_02
```

```python
mflx_labor_02
```

```python
mflx_kid_02['trimmed_folio'] = mflx_kid_02['folio'].apply(lambda x: int(str(int(x))[:-3]) if not pd.isna(x) else np.nan)
mflx_kid_05['trimmed_folio'] = mflx_kid_05['folio'].apply(lambda x: int(str(x)[:-3]) if not pd.isna(x) else np.nan)
mflx_kid_09['trimmed_folio'] = mflx_kid_09['folio'].apply(lambda x: int(str(x)[:-5]) if not pd.isna(x) else np.nan)
mflx_kid_02
```

```python
treatment_ls = []
household_income_ls = []
household_head_secondary_ls = []
household_children_ls = []
city_bigger_15k_ls = []

for index, row in mflx_kid_02.iterrows():
  trimmed_folio = row["trimmed_folio"]
  head_household_rows = mflx_labor_02[(mflx_labor_02['trimmed_folio'] == trimmed_folio) & (mflx_labor_02['ls05_1'] == 1)]
  if len(head_household_rows) > 0:
    treatment = head_household_rows.iloc[0]["treatment"]
    household_income = head_household_rows.iloc[0]["household_income"]
    household_head_secondary = head_household_rows.iloc[0]["household_head_secondary"]
    household_children = head_household_rows.iloc[0]["household_children"]
    city_bigger_15k = head_household_rows.iloc[0]["city_bigger_15k"]
    
    treatment_ls.append(treatment)
    household_income_ls.append(household_income)
    household_head_secondary_ls.append(household_head_secondary)
    household_children_ls.append(household_children)
    city_bigger_15k_ls.append(city_bigger_15k)
  else:
    treatment_ls.append(np.NaN)
    household_income_ls.append(np.NaN)
    household_head_secondary_ls.append(np.NaN)
    household_children_ls.append(np.NaN)
    city_bigger_15k_ls.append(np.NaN)

mflx_kid_02["treatment"] = treatment_ls
mflx_kid_02["household_income"] = household_income_ls
mflx_kid_02["household_head_secondary"] = household_head_secondary_ls
mflx_kid_02["household_children"] = household_children_ls
mflx_kid_02["city_bigger_15k"] = city_bigger_15k_ls
```

```python
mflx_kid_02.dropna(subset=['treatment'], inplace=True)
mflx_kid_05.dropna(subset=['treatment'], inplace=True)
mflx_kid_09.dropna(subset=['treatment'], inplace=True)
```

```python
mflx_kid_02["period"] = 1
mflx_kid_05["period"] = 2
mflx_kid_09["period"] = 3
```

```python
tracked = {}
counter = 1

#assigns unique integer ID to each folio-ls pair
for index, row in mflx_kid_02.iterrows():
  folio = int(row['trimmed_folio'])
  id = int(row['ls'])
  if folio in tracked:
    if id not in tracked[folio]:
      tracked[folio][id] = counter
      counter += 1
  elif folio not in tracked:
    tracked[folio] = {}
    tracked[folio][id] = counter
    counter += 1

for index, row in mflx_kid_05.iterrows():
  folio = int(row['trimmed_folio'])
  id = int(row['ls'])
  if folio in tracked:
    if id not in tracked[folio]:
      tracked[folio][id] = counter
      counter += 1
  elif folio not in tracked:
    tracked[folio] = {}
    tracked[folio][id] = counter
    counter += 1

for index, row in mflx_kid_09.iterrows():
  folio = int(row['trimmed_folio'])
  id = int(row['ls'])
  if folio in tracked:
    if id not in tracked[folio]:
      tracked[folio][id] = counter
      counter += 1
  elif folio not in tracked:
    tracked[folio] = {}
    tracked[folio][id] = counter
    counter += 1

#add unique id column
unique_id_02_ls = []
for index, row in mflx_kid_02.iterrows():
  folio = int(row['trimmed_folio'])
  id = int(row['ls'])
  unique_id_02_ls.append(tracked[folio][id])
mflx_kid_02["unique_id"] = unique_id_02_ls

unique_id_05_ls = []
for index, row in mflx_kid_05.iterrows():
  folio = int(row['trimmed_folio'])
  id = int(row['ls'])
  unique_id_05_ls.append(tracked[folio][id])
mflx_kid_05["unique_id"] = unique_id_05_ls

unique_id_09_ls = []
for index, row in mflx_kid_09.iterrows():
  folio = int(row['trimmed_folio'])
  id = int(row['ls'])
  unique_id_09_ls.append(tracked[folio][id])
mflx_kid_09["unique_id"] = unique_id_09_ls
```

```python
age_ls = []
for index, row in mflx_kid_02.iterrows():
  trimmed_folio = row["trimmed_folio"]
  ls = row["ls"]
  matching_rows = mflx_labor_02[(mflx_labor_02['trimmed_folio'] == trimmed_folio) & (mflx_labor_02['ls_x'] == ls)]
  if len(matching_rows) > 0:
    age = matching_rows.iloc[0]["age_yrs"]
    age_ls.append(age)
  else:
    age_ls.append(np.NaN)

mflx_kid_02["age_yrs"] = age_ls

age_ls = []
for index, row in mflx_kid_05.iterrows():
  trimmed_folio = row["trimmed_folio"]
  if pd.notna(row["ls"]):
    ls = int(row["ls"])
  else:
    ls = row["ls"]
  matching_rows = mflx_labor_02[(mflx_labor_02['trimmed_folio'] == trimmed_folio) & (mflx_labor_02['ls_x'] == ls)]
  if len(matching_rows) > 0:
    age = matching_rows.iloc[0]["age_yrs"]
    age_ls.append(age)
  else:
    age_ls.append(np.NaN)

mflx_kid_05["age_yrs"] = age_ls

age_ls = []
for index, row in mflx_kid_09.iterrows():
  trimmed_folio = row["trimmed_folio"]
  if pd.notna(row["ls"]):
    ls = int(row["ls"])
  else:
    ls = row["ls"]
  matching_rows = mflx_labor_02[(mflx_labor_02['trimmed_folio'] == trimmed_folio) & (mflx_labor_02['ls_x'] == ls)]
  if len(matching_rows) > 0:
    age = matching_rows.iloc[0]["age_yrs"]
    age_ls.append(age)
  else:
    age_ls.append(np.NaN)

mflx_kid_09["age_yrs"] = age_ls
```
